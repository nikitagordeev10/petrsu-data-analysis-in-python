На двух метеостанциях в одном из городов России за один период времени собирали информацию о погоде.  Написать функцию weather_forecast(day), на вход которой подается номер дня day (число: 1, 2, …, 10). В этой функции реализовать:

- Создание словаря weather_station1 с полученной с первой метеостанции информацией о погоде, при этом ключ = номер дня (1, 2, … 10), значение = [температура, атмосферное давление]: 

1: [20, 735], 

2: [21, 737], 

3: [23, 740], 

4: [20, 739], 

5: [21, 750],

6: [21, 751], 

7: [22, 749], 

8: [19, 745], 

9: [22, 752],

10: [23, 753].

- Создание словаря weather_station2 с полученной со второй метеостанции информацией о погоде, при этом ключ = номер дня (1, 2, … 10), значение = [температура, влажность]:  

1: [20, 89], 

2: [20, 90], 

3: [24, 95], 

4: [20, 88], 

5: [21, 87],

6: [21, 80], 

7: [22, 72], 

8: [18, 80], 

9: [22, 75],

10: [23, 78].

Функция weather_forecast(day) должна находить температуру, атмосферное давление, влажность в указанный день day, при этом температура ищется как среднее арифметическое значений температуры, которые получены на weather_station1 и weather_station2. С помощью return функция weather_forecast(day) должна возвращать сумму полученных температуры, атмосферного давления, влажности.

Для примера:

| Тест                | Результат |
|---------------------|-----------|
| print(weather_forecast(1))  | 844.0        |
| print(weather_forecast(4)) | 847.0       |

```python
def weather_forecast(day):
    # Данные с первой метеостанции: ключ - день, значение - [температура, давление]
    weather_station1 = {
        1: [20, 735],
        2: [21, 737],
        3: [23, 740],
        4: [20, 739],
        5: [21, 750],
        6: [21, 751],
        7: [22, 749],
        8: [19, 745],
        9: [22, 752],
        10: [23, 753]
    }

    # Данные со второй метеостанции: ключ - день, значение - [температура, влажность]
    weather_station2 = {
        1: [20, 89],
        2: [20, 90],
        3: [24, 95],
        4: [20, 88],
        5: [21, 87],
        6: [21, 80],
        7: [22, 72],
        8: [18, 80],
        9: [22, 75],
        10: [23, 78]
    }
    
    # Извлекаем температуру и давление с первой метеостанции для заданного дня
    temp1, pressure = weather_station1[day]    
    
    # Извлекаем температуру и влажность со второй метеостанции для заданного дня
    temp2, humidity = weather_station2[day]    
    
    # Вычисляем среднюю температуру между двумя метеостанциями
    avg_temp = (temp1 + temp2) / 2
    
    # Возвращаем сумму средней температуры, давления и влажности
    return avg_temp + pressure + humidity
```

```python
print(weather_forecast(1))
```

```python
print(weather_forecast(4))
```
